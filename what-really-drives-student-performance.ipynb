```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/students-performance-in-exams/StudentsPerformance.csv")
df.head()
```

```python
df["average_score"] = (
    df["math score"] + df["reading score"] + df["writing score"]
) / 3

df[["math score", "reading score", "writing score", "average_score"]].head()
```

```python
df.groupby("test preparation course")["average_score"].mean().plot(
    kind="bar", title="Test Preparation vs Average Score"
)
plt.ylabel("Average Score")
plt.show()
```

Insight: Students who completed the test preparation course achieved noticeably higher average scores, showing that structured preparation has a strong positive impact on performance.
